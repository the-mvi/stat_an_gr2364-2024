## Основные теоретические положения


Случайный вектор $\vec{\xi} = \left(\xi_1, \xi_2, ..., \xi_n \right)^{\intercal} : \Omega \rightarrow \mathbb{R}^n $ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\vec{\zeta} = \left(\zeta_1, ..., \zeta_m \right)^T $, вещественный вектор $\vec{\mu} = \left(\mu_1, ... , \mu_n \right)^{\intercal} $ и нулевая матрица $ A $ размерности $n \times m$, такие, что $\vec{\xi} = A\vec{\zeta} + \mu $

Вектор математических ожиданий:

$\mu = \mathbb(E){\vec{\xi}} = (\mathbb(E)\xi_1, ..., \mathbb(E)\xi_n)^{\intercal} $

Матрица ковариаций:

$\Sigma = (\sigma_{ij})_{1 \leq ij \leq n}, \sigma_{ij} = cov(\xi_i, \xi_j)$

Плотность нормального случайного вектора:

$ p_{\vec{\xi}}(x_1, .., x_n) = \frac{1}{(2\pi)^{n/2}\sqrt{det(\Sigma)}}exp(-\frac{1}{2}(\vec{x} - \mu)^{\intercal} \Sigma^{-1}(\vec{x}-\mu)) $

Афинное преобразование:

Преобразование, которое переводит вектор в стандартный нормальный.
