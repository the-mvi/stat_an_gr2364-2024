## Основные теоретические положения

**Многомерное нормальное распределение**

Случайный вектор $\overrightarrow{\xi} = \left(\xi_1,\dots , \xi_n\right)^T$ имеет многомерное нормальное распределение, если $\exists$ вектор независимых стандартных случайных вероятностей $\overrightarrow{\zeta}=\left(\zeta_1, \dots, \zeta_n\right)^T$, вещественный вектор $\overrightarrow{\mu}=\left(\mu_1, \dots, \mu_n\right)^T$ и ненулевая матрица $A$ размерности $m\times n$, такие, что:

$\overrightarrow{\xi} = A \overrightarrow{\zeta} + \overrightarrow{\mu}$
    
$\overrightarrow{\xi} \sim \mathcal{N} \left(\mu,\Sigma\right)$
    
$\overrightarrow{\mu} = \overrightarrow{\mathbb{E}_{\xi}} = \left(\mathbb{E}_{\xi_1}, \dots, \mathbb{E}_{\xi_n}\right)^T$ - вектор математических ожиданий
    
$\Sigma = \left(\sigma_{ij}\right)_{i \leq i,j \leq n}, \sigma_{ij} = \mathrm{cov}(\xi_i, \xi_j)$ - матрица ковариаций
    
**Плотность нормального случайного вектора**
    
$p_{\overrightarrow{\xi}}(x_1, \dots, x_n)=\cfrac{1}{\left(2 \pi\right)^{n/2} \sqrt{det \left(\Sigma\right)}} \cdot \exp\left(-\cfrac{1}{2}\left(\overrightarrow{x} - \mu\right)^T\Sigma^{-1}\left(\overrightarrow{x} - \mu\right)\right)$
    
$\overrightarrow{\xi} \sim \mathcal{N}\left(\mu, \Sigma\right)$
    
$\overrightarrow{\eta}=A \overrightarrow{\xi}+b$
  
$\mathbb{E} \overrightarrow{\eta}=A \mu+b$
$\mathrm{Var}\left(\overrightarrow{\eta}\right)=A \Sigma A^{\intercal}$

### Плотность двумерного нормального распределения имеет вид:

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p_{{\xi, \eta}}$.

$3x^2 - 3xy + 7y^2 + 3x + 11y + 7 = 3\left( x - \cfrac{y}{2} \right)^2 + 6\cfrac{1}{4}y^2 + 3x + 11y + 7 = 3\left(x - \cfrac{y}{2}\right)^2 + 3\left( x - \cfrac{y}{2} \right) + 6\cfrac{1}{4}y^2 + 12.5y + 7 = 3\left( \left( x - \cfrac{y}{2} \right)^2 + \left( x - \cfrac{y}{2} + \cfrac{1}{4} \right) \right) + \cfrac{25}{4}y^2 + 12.5y + 6.25 = $

$ = 3 \left( \left( x - \cfrac{y}{2} \right) + \cfrac{1}{2} \right)^2 + \cfrac{25}{4}\left( y + 1 \right)^2 = ... = 3\left( x + 1 \right)^2 + 2 \cdot \left(-\cfrac{3}{2}\right)\left( x + 1 \right)\left( y + 1 \right) + 7\left( y + 1 \right)^2$

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix}=\begin{pmatrix} -1 \\ -1 \end{pmatrix}$

$\Sigma^{-1} =\begin{pmatrix} 3 & -\cfrac{3}{2} \\ -\cfrac{3}{2} & 7 \end{pmatrix}$

$\Sigma = \cfrac{4}{75}\begin{pmatrix} 7 & \cfrac{3}{2} \\ \cfrac{3}{2} & 3 \end{pmatrix} = \begin{pmatrix} \cfrac{28}{75} & \cfrac{2}{25} \\ \cfrac{2}{25} & \cfrac{4}{25} \end{pmatrix} = \begin{pmatrix} 0.373 & 0.08 \\ 0.08 & 0.16 \end{pmatrix}$

$\mathbb{D}\xi = \cfrac{28}{75} \text{;} \ \mathbb{D}\eta = \cfrac{4}{25}$

$\mathrm{cov}(\xi, \eta) = \cfrac{2}{25}$

$\rho(\xi, \eta) = \cfrac{\mathrm{cov}(\xi, \eta)}{\sqrt{\mathbb{D}\xi \cdot \mathbb{D}\eta}} = \cfrac{2}{25} \cdot \cfrac{1}{\sqrt{\cfrac{28}{75} \cdot \cfrac{4}{25}}} = \cfrac{\sqrt{21}}{14}$

$C = \cfrac{1}{(2\pi)^{n/2} \cdot \sqrt{det \Sigma}} = \cfrac{1}{(2\pi)^{2/2} \cdot \sqrt{\cfrac{4}{75}}} = \cfrac{5\sqrt{3}}{4\pi}$

$p_{{\xi, \eta}}(x, y)=\cfrac{5\sqrt{3}}{4\pi} \cdot \exp \left(3 \left( \left( x - \cfrac{y}{2} \right) + \cfrac{1}{2} \right)^2 + \cfrac{25}{4}\left( y + 1 \right)^2\right)$;

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, e

x, y = np.meshgrid(np.arange(-5, 5, 0.1), np.arange(-5, 5, 0.1))
z = 5 * 3 ** 0.5 / (4 * pi) * e ** (-0.5 * (3*x**2 - 3*x*y + 7*y**2 + 3*x + 11*y + 7))
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(x, y, z, color='green')
plt.show()
```

### 2. Найти аффинное преобразование, переводящее исходный случайный вектор в стандартный нормальный.

$ q(x, y) = 3 \left( \left( x - \cfrac{y}{2} \right) + \cfrac{1}{2} \right)^2 + \cfrac{25}{4}\left( y + 1 \right)^2 = \left( \sqrt{3} \left( x - \cfrac{y}{2} \right) + \cfrac{\sqrt{3}}{2} \right)^2 + \left( \cfrac{5}{2}y + \cfrac{5}{2} \right)^2$

$ \begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = A \begin{pmatrix} \xi \\ \eta \end{pmatrix} + \beta  = \begin{pmatrix} \sqrt{3} & -\frac{\sqrt{3}}{2} \\ 0 & \cfrac{5}{2} \end{pmatrix} \begin{pmatrix} \xi \\ \eta \end{pmatrix} + \begin{pmatrix} \cfrac{\sqrt{3}}{2} \\ \cfrac{5}{2} \end{pmatrix}$

$ \mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix}= \begin{pmatrix} \sqrt{3} & -\frac{\sqrt{3}}{2} \\ 0 & \cfrac{5}{2} \end{pmatrix} \begin{pmatrix} -1 \\ -1 \end{pmatrix} + \begin{pmatrix} \cfrac{\sqrt{3}}{2} \\ \cfrac{5}{2} \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

Проверка:

### 3. Найти ортогональное преобразовние, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

$\Sigma^{-1} =\begin{pmatrix} 3 \ -\cfrac{3}{2} \\ -\cfrac{3}{2} \ 7 \end{pmatrix}$

$det \left( \Sigma^{-1} - \lambda E \right) = 0$

$det \begin{pmatrix} 3 - \lambda & -\cfrac{3}{2} \\ -\cfrac{3}{2} & 7 - \lambda \end{pmatrix} = \left( 3 - \lambda \right)\left( 7 - \lambda \right) - \left(-\cfrac{3}{2}\right) \cdot \left(-\cfrac{3}{2}\right) = \lambda^2 - 10\lambda + \cfrac{75}{4} = 0$

$\lambda_{1} = \cfrac{5}{2} ;\ \lambda_{2} = \cfrac{15}{2}$

$V_{1} = \begin{pmatrix} 3 - \cfrac{5}{2} & -\cfrac{3}{2} \\ -\cfrac{3}{2} & 7 - \cfrac{5}{2} \end{pmatrix} = \begin{pmatrix} \cfrac{1}{2} & -\cfrac{3}{2} \\ -\cfrac{3}{2} & \cfrac{9}{2} \end{pmatrix}$

$V_{2} = \begin{pmatrix} 3 - \cfrac{15}{2} & -\cfrac{3}{2} \\ -\cfrac{3}{2} & 7 - \cfrac{15}{2} \end{pmatrix} = \begin{pmatrix} -\cfrac{9}{2} & -\cfrac{3}{2} \\ -\cfrac{3}{2} & -\cfrac{1}{2} \end{pmatrix}$

1) $\begin{cases} \cfrac{1}{2}x -\cfrac{3}{2}y = 0 \\ -\cfrac{3}{2}x + \cfrac{9}{2}y = 0 \end{cases} \rightarrow \overline{V_{1}} = \alpha \begin{pmatrix} 3 \\ 1 \end{pmatrix}$

$\alpha = \cfrac{1}{\sqrt{1^2 + 3^2}} = \cfrac{1}{\sqrt{10}}$

2) $\begin{cases} -\cfrac{9}{2}x -\cfrac{3}{2}y = 0 \\ -\cfrac{3}{2}x - \cfrac{1}{2}y = 0 \end{cases} \rightarrow \overline{V_{1}} = \beta \begin{pmatrix} -1 \\ 3 \end{pmatrix}$

$\alpha = \cfrac{1}{\sqrt{(-1)^2 + 3^2}} = \cfrac{1}{\sqrt{10}}$

$Q^T = \cfrac{1}{\sqrt{10}} \begin{pmatrix} 3 & -1 \\ 1 & 3 \end{pmatrix} \rightarrow Q = \cfrac{1}{\sqrt{10}} \begin{pmatrix} 3 & 1 \\ -1 & 3 \end{pmatrix}$

$\Sigma_{id} = Q \Sigma Q^T = \begin{pmatrix} \cfrac{1}{\lambda_{1}} & 0 \\ 0 & \cfrac{1}{\lambda_{2}} \end{pmatrix} = \begin{pmatrix} \cfrac{2}{5} & 0 \\ 0 & \cfrac{2}{15} \end{pmatrix}$

Проверка:

$\cfrac{1}{\sqrt{10}} \begin{pmatrix} 3 & -1 \\ 1 & 3 \end{pmatrix} \cdot \cfrac{4}{75}\begin{pmatrix} 3 & \cfrac{3}{2} \\ \cfrac{3}{2} & 7 \end{pmatrix} \cdot \cfrac{1}{\sqrt{10}} \begin{pmatrix} 3 & 1 \\ -1 & 3 \end{pmatrix} = \begin{pmatrix} \cfrac{2}{15} & 0 \\ 0 & \cfrac{2}{5} \end{pmatrix}$

Все верно, т.к. на главной диагонали обратные собственные числа.

$ det(\Sigma_{id}) = \cfrac{2}{5} \cdot \cfrac{2}{15} = \cfrac{4}{75}$ 

Коэфф. перед экспонентой в плотности: $ K = \cfrac{1}{\sqrt{(2\pi)^2 \cdot \cfrac{4}{75}}} = \cfrac{5\sqrt{3}}{8\pi}$

$p_{\xi, \eta}(x, y) = \cfrac{5\sqrt{3}}{8\pi} \cdot \exp\left(-\cfrac{1}{2}\left(2.5x^2 + 7.5y^2\right)\right)$

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, e

x, y = np.meshgrid(np.arange(-3, 3, 0.2), np.arange(-3, 3, 0.2))
z = 5 * 3 ** 0.5 / (8 * pi) * e ** (-0.5 * (2.5 * x**2 + 7.5 * y**2))
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(x, y, z, color='green')
plt.show()
```

### 4. Вычислить характеристики совместного распределения случайного вектора $\left( -5\xi - \eta, -2\xi + 3 \eta \right)^T$ и записать его плотность. Построить график плотности полученного распределения.

   $ \vec{z}= \begin{pmatrix} -5\xi - \eta \\ -2\xi + 3\eta \end{pmatrix} =  \begin{pmatrix} -5 & -1 \\ -2 & 3 \end{pmatrix} \begin{pmatrix} \xi \\ \eta \end{pmatrix} + \begin{pmatrix} 0 \\ 0 \end{pmatrix}$

   $ \mathbb{E}_z = \begin{pmatrix} -5 & -1 \\ -2 & 3 \end{pmatrix} \begin{pmatrix} -1 \\ -1 \end{pmatrix} = \begin{pmatrix} 6 \\ -1 \end{pmatrix}$

   $ \mathbb{D}_z = \cfrac{4}{75} \begin{pmatrix} -5 & -1 \\ -2 & 3 \end{pmatrix} \begin{pmatrix} 7 & \cfrac{3}{2} \\ \cfrac{3}{2} & 3 \end{pmatrix} \begin{pmatrix} -5 & -2 \\ -1 & 3 \end{pmatrix} = \cfrac{4}{75} \begin{pmatrix} 193 & \cfrac{83}{2} \\ \cfrac{83}{2} & 37 \end{pmatrix}$

   $ \Sigma^{-1} = \cfrac{1}{289} \begin{pmatrix} 37 & -\cfrac{83}{2} \\ -\cfrac{83}{2} & 193 \end{pmatrix}$ 

   $ K = \cfrac{5\sqrt{3}}{17 \cdot 2 \cdot 2 \cdot \pi} = \cfrac{5\sqrt{3}}{68\pi}$ 
   
   $ \rho_z(x,y)= \cfrac{5\sqrt{3}}{68\pi} \cdot \exp\left(-\cfrac{1}{2}\left(\cfrac{37}{289}(x-6)^2 - \cfrac{83}{289}(x-6)(y+1) + \cfrac{193}{289}(y+1)^2\right)\right)$

   
   

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi, e

x, y = np.meshgrid(np.arange(-10, 10, 0.2), np.arange(-10, 10, 0.2))
z = (5 * 3**(0.5) ) / (68 * pi) * e ** (-0.5 * (37/289 * (x - 6)**2 - 83/289 * (x - 6)*(y + 1) + 193/289 * (y + 1)**2))
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.plot_surface(x, y, z, color='green')
plt.show()
```

### 5. Найти условное распределение $\xi$ при условии $\eta$. Вычислить $\mathbb{E}\left(\xi|\eta\right)$ и $\mathbb{D}\left(\xi|\eta\right)$.

$ \mathbb{E}\left( \xi|\eta \right) = -1 + \cfrac{\cfrac{2}{25}}{\cfrac{4}{25}}\left(\eta + 1\right) = \cfrac{1}{2}\eta - \cfrac{1}{2} $

$\mathbb{D}\left( \xi | \eta \right) = \cfrac{28}{75} - \cfrac{\left(\cfrac{2}{25}\right)^2}{\cfrac{4}{25}} = \cfrac{1}{3}$

$ \rho_{\xi}(\xi|\eta) \sim \mathcal{N}(\cfrac{1}{2}\left(\eta - 1\right), \cfrac{1}{3}) $
