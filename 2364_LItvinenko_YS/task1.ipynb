$\rho (\xi, \eta) = \cfrac{\mathrm{cov} (\xi, \eta)}{\sqrt{\mathbb{D} \xi \cdot \mathbb{D} \eta}}$  

Формула для ортагонального преобразования:  

$\Sigma_{id} = Q \cdot \Sigma \cdot Q^{T} \rightarrow \Sigma_{id}$ 

Формула для нахождения условного распределения:

$\rho_{\xi|\eta}(x)= \frac{\rho_{\xi,\eta}(x,y)}{\rho_\eta(y)}$

# Задача №1.

Плотность двумерного нормального распределения имеет вид:

$ p_{\xi, \eta}  (x, y) = C \cdot \exp \left(-\cfrac{1}{2}\left(2x^2+4xy+5y^2-8x-2y+11\right)\right) $

Необходимо вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $ p_{\xi, \eta} $ . 

$ С= \frac{1}{(2\pi)^2 \cdot \sqrt{det\sum_{}^{}}}   $

Для того чтобы найти коэфицент C необходимо преобразоать исходное выражение:
$2x^2+4xy+5y^2-8x-2y+11=2x^2+4xy+5y^2-8x-2y+11-8y+8y=2x^2+4xy-8x+2y^2-8y+8+3y^2+6y+3=$
$=2x^2+4xy-8x+2y^2-8y+8+3(y^2+2y+1)=2x^2+4xy-8x+2y^2-8y+8+3(y+1)^2=$
$=2x^2+4xy-8x+2(y-2)^2+3(y+1)^2=2(x^2+2xy-4x+(y-2)^2)+3(y+1)^2=$ 
$=2(x^2+2x(y-2)+(y-2)^2)+3(y+1)^2=2(x+(y-2))^2+3(y+1)^2=$
$=2((x-3)+(y+1))^2+3(y+1)^2=2((x-3)^2+2(x-3)(y+1)+(y+1)^2)+3(y+1)^2=$
$=2(x-3)^2+4(x-3)(y+1)+5(y+1)^2$

После преобразования видно, что вектор математических ожиданий будет равен:  

$E=\left( \begin{matrix}
3   \\
-1  
\end{matrix}\right)$

Ковариационная матрица в степени -1 будет равна:  

$ \Sigma^{-1} = \begin{pmatrix}
2 & \ 2\\
2 & 5
\end{pmatrix} $  

Найдем обратную матрицу:  

$ \Sigma =\frac{1}{6} \begin{pmatrix}
5 &-2\\
-2 & 2
\end{pmatrix} $ 

После нахождения ковариационной матрицы необходимо найти дисперсию и ковариацию:  

$ cov(\xi, \eta)=-\frac{2}{6}=-\frac{1}{3}$

$\mathbb{D} \xi=\frac{5}{6}$ 

$\mathbb{D} \eta=\frac{2}{6}=\frac{1}{3}$  

Зная дисперсии и ковариацию по формуле найдем коэфицент кореляции:  

$ \rho (\xi, \eta) = \cfrac{\mathrm{cov} (\xi, \eta)}{\sqrt{\mathbb{D} \xi \cdot \mathbb{D} \eta}} = -\cfrac{-\frac{1}{3}}{\sqrt{\frac{5}{6} \cdot \frac{1}{3}}} = -\frac{1}{3} \cdot \sqrt{\frac{5}{18}}=-\frac{\sqrt{2}}{5} $  

Теперь по формуле вычислим коэфицент C:

$ C = \frac{1}{\left(2\pi\right)^{n/2} \sqrt{\det\Sigma}} = \frac{1}{\left(2\pi\right)^{2/2} \sqrt{\frac{1}{6}}} =\frac{\sqrt{6}}{2\pi} $

После нахождения коэфицента C подставим его в изначальное выражение:  

$ p_{\xi, \eta}  (x, y) = \frac{\sqrt{6}}{2\pi} \cdot \exp \left(-\cfrac{1}{2}\left(2x^2+4xy+5y^2-8x-2y+11\right)\right) $  

Построим график функции:



```python
import numpy as np
import matplotlib.pyplot as plt

from math import pi, e, sqrt

x = np.arange(-2, 2, 0.05)
y = np.arange(-2, 2, 0.05)
xgrid, ygrid = np.meshgrid(x, y)
zgrid =  sqrt(6) / (2 * pi) * e ** (-0.5 * (2 * xgrid ** 2 + 4 * xgrid * ygrid + 5 * ygrid ** 2 - 8 * xgrid - 2 * ygrid + 11))
fig = plt.figure()
ax = plt.axes(projection="3d")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.plot_wireframe(xgrid, ygrid, zgrid)
plt.show()
```
