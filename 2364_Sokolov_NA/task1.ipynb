## Основные теоретические положения

Случайный вектор $\vec\xi=\left(\xi_1,\dots,\xi_n\right)^\intercal\colon\Omega\to\mathbb{R}^n$ имеет многомерное нормальное распределение ($\vec\xi\sim\mathcal{N}\left(\mu,\Sigma\right)$), если существует вектор независимых стандартных нормальных случайных величин $\vec\zeta=\left(\zeta_1,\dots,\zeta_n\right)^\intercal$, вещественный вектор $\mu=\left(\mu_1,\dots,\mu_n\right)^\intercal$ и ненулевая матрица $A$ размерности $n\times m$, такие что

$$
\vec\xi=A\vec\zeta+\mu.
$$

$\mu=\mathbb{E}\vec\xi=(\mathbb{E}\xi_1,\dots,\mathbb{E}\xi_n)^\intercal$ - вектор математических ожиданий.

$\Sigma=\left(\sigma_{ij}\right)_{1\leqslant i,j\leqslant n}$, $\sigma_{ij}=\mathrm{cov}\left(\xi_i,\xi_j\right)$ - матрица ковариаций.

Плотность случайного многомерного нормального вектора:

$$
p_{\vec\xi}\left(\vec x\right)=\cfrac{1}{(2\pi)^{n/2}\sqrt{\mathrm{det}\Sigma}}\mathrm{exp}\left(-\cfrac{1}{2}(\vec x-\mu)^\intercal\Sigma^{-1}(\vec x - \mu)\right),
$$

$\vec x=\left(x_1,\dots, x_n\right)^\intercal$.
